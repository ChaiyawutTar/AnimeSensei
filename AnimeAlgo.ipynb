```python
text = "Koyoharu Gotouge : Original Creator, Haruo Sotozaki : Director, Akira Matsushima : Character Design, Aimer : Song Performance"
text2 = 'Mo Xiang Tong Xiu : Original Creator, Xiong Ke : Chief Director, Ma Chendi : Chief Director, Sun Yujing : Music, Weng Teng : Music, Feng Shuo : Music, Shen Lin : Character Design & Chief Animation Director, Liang Sha : Screenplay'
def staff_to_dict(text):
    my_list = []
    
# Split the text by comma
    items = text.split(',')

# Iterate over each item
    for item in items:
        # Split the item by colon
        parts = item.split(':')
        # Strip any whitespace from the parts
        key = parts[1].strip()
        value = parts[0].strip()
        # Add the key-value pair to the dictionary
        my_dict = {key: value}
        # Append the dictionary to the list
        my_list.append(my_dict)
    return str(my_list)
staff_to_dict(text)
staff_to_dict(text2)
# df['new_staff'] = df['staff'].apply(staff_to_dict)
```

```python
def extract_director(staff_str):
    if isinstance(staff_str, str):
        staff_list = staff_str.split(', ')
        for staff_member in staff_list:
            if 'Director' in staff_member:
                return staff_member.split(':')[0].strip()
    return None

# Create a new column 'director' based on the 'staff' column
df['director'] = df['staff'].apply(extract_director)
df
```

```python
temp_data = {'Name':df['Name'].dropna().to_list(), 'director':df['director'].dropna().to_list()}
temp_data
def is_path_of_length_3_or_less(char, director, df):
    # Create a graph from the dataframe
    G = nx.Graph()
    G = nx.from_pandas_edgelist(df, source='Name', target='director')
    
    # Check if there is a path of length 3 or less between char1 and char2
    path = nx.shortest_path(G, source=char, target=director, weight=None)
    if len(path) <= 4:
        return True
    else:
        return False
is_path_of_length_3_or_less('Demon Slayer: Kimetsu no Yaiba - Entertainment District Arc', 'Haruo Sotozaki', temp_data)
```

```python
# Create a new graph object
G = nx.Graph()

# Add nodes to the graph
for title in df['Name']:
    G.add_node(title)

# Add edges to the graph
for index, row in df.iterrows():
    related_anime = row['Related_anime']
    if isinstance(related_anime, str) and related_anime != '':
        related_anime_list = related_anime.split(', ')
        for anime in related_anime_list:
            G.add_edge(row['Name'], anime)

# Define a function to find the shortest path between two anime titles
def find_shortest_path(G, start, end):
    path = nx.dijkstra_path(G, start, end)
    return path

# Find the shortest path between "Attack on Titan" and "Fullmetal Alchemist: Brotherhood"
shortest_path = find_shortest_path(G, "Jujutsu Kaisen", "Attack on Titan The Final Season: Part II")
print(shortest_path)

```
